```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
#load data

df=pd.read_csv('usa.csv', encoding='utf8')

df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)

df.info()
```

```python
df['City'].unique()
```

```python
x= df[['Beds','Baths','Living Space', 'City', 'Zip Code Density' ]]  #features
y=df['Price'] #label

x.head()
```

```python
#encode categorical data
from sklearn.preprocessing import LabelEncoder
city_encoder= LabelEncoder()
city_encoder.fit(x['City']) #fit the encoder to the unique values in the 'City' column

x['City']=city_encoder.transform(x['City']) #transform the 'City' column using the fitted encoder

x.head(20)
```

```python
y.head()
```

```python
x.info()
```

```python
#split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


print(x_train.shape)
print(x_test.shape)

```

```python
#linear regression - training the model
from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(x_train, y_train)

```

```python
#evaluate the model
y_pred=model.predict(x_test)

from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error, mean_absolute_error

mse= mean_squared_error(y_test, y_pred)
rmse= np.sqrt(mse)

r2= r2_score(y_test, y_pred)

mae= mean_absolute_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared: {r2}')
print(f'Mean Absolute Error: {mae}')
```

```python
#other regression models
#random forest regressor
#svr regression

from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)

rf_y_pred = rf_model.predict(x_test)


mse= mean_squared_error(y_test, rf_y_pred)
rmse= np.sqrt(mse)

r2= r2_score(y_test, rf_y_pred)

mae= mean_absolute_error(y_test, rf_y_pred)
print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared: {r2}')
print(f'Mean Absolute Error: {mae}')
```

```python
#svr regression
from sklearn.svm import SVR
svr_model = SVR(kernel='linear')
svr_model.fit(x_train, y_train)
svr_y_pred = svr_model.predict(x_test)


mse= mean_squared_error(y_test, svr_y_pred)
rmse= np.sqrt(mse)

r2= r2_score(y_test, svr_y_pred)

mae= mean_absolute_error(y_test, svr_y_pred)
print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared: {r2}')
print(f'Mean Absolute Error: {mae}')
```

```python
#tune model - grid search
from sklearn.model_selection import GridSearchCV
grid={
    'n_estimators': [100],
    'max_depth': [3,4,5],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 3]
}



rf_grid_search = GridSearchCV(estimator=rf_model, param_grid=grid, cv=3, n_jobs=-1,verbose=2)
rf_grid_search.fit(x_train, y_train)
print(f'Best parameters: {rf_grid_search.best_params_}')
```

```python
best_rf_model = rf_grid_search.best_estimator_
best_rf_y_pred = best_rf_model.predict(x_test)

mse= mean_squared_error(y_test, best_rf_y_pred)
rmse= np.sqrt(mse)
r2= r2_score(y_test, best_rf_y_pred)
mae= mean_absolute_error(y_test, best_rf_y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared: {r2}')
print(f'Mean Absolute Error: {mae}')
```
