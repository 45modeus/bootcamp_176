```python
df2['mainroad'] = df2['mainroad'].map({'yes': 1, 'no': 0})
df2['guestroom'] = df2['guestroom'].map({'yes': 1, 'no': 0})
df2['basement'] = df2['basement'].map({'yes': 1, 'no': 0})
df2['hotwaterheating'] = df2['hotwaterheating'].map({'yes': 1, 'no': 0})
df2['airconditioning'] = df2['airconditioning'].map({'yes': 1, 'no': 0})
df2['prefarea'] = df2['prefarea'].map({'yes': 1, 'no': 0})


# df2['mainroad'] = df2['mainroad'].astype(bool).astype(int)
```

```python
df2.head()
```

```python
#predicting price of a house based on all columns


#preprocessing
furnishing_encoder= {'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0}
df2['furnishingstatus'] = df2['furnishingstatus'].map(furnishing_encoder)

df2['furnishingstatus']
```

```python
df2.head()
```

```python
x= df2.drop(['price'], axis=1)  # Features
y= df2['price']  # Target variable

x.head()
```

```python
y.head()
```

```python
#split dataser into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
results={}
```

```python
#linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train) #train the model

#evaluate the model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse= np.sqrt(mse)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")

results['Linear Regression'] = {
    'Mean Squared Error': mse,
    'R-squared': r2,
    'Mean Absolute Error': mae,
    'Root Mean Squared Error': rmse
}
```

```python
#decision tree regressor
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train, y_train)  # Train the model

# Evaluate the model

y_pred = dt_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse= np.sqrt(mse)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")

results['Decision Tree Regressor'] = {
    'Mean Squared Error': mse,
    'R-squared': r2,
    'Mean Absolute Error': mae,
    'Root Mean Squared Error': rmse
}
```

```python
#random forest regressor
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)  # Train the model

# Evaluate the model
y_pred = rf_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse= np.sqrt(mse)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")

results['Random Forest Regressor'] = {
    'Mean Squared Error': mse,
    'R-squared': r2,
    'Mean Absolute Error': mae,
    'Root Mean Squared Error': rmse
}
```

```python
#print params of rf model
print("Random Forest Regressor Parameters:")
print(rf_model.get_params())
```

```python
#naive bayes regressor
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)  # Train the model

# Evaluate the model

y_pred = nb_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse= np.sqrt(mse)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")

results['Naive Bayes Regressor'] = {
    'Mean Squared Error': mse,
    'R-squared': r2,
    'Mean Absolute Error': mae,
    'Root Mean Squared Error': rmse
}
```

```python
#tuning model
#selected the linear regression model for further tuning

param_grid= {
    'fit_intercept': [True, False],
    'normalize': [True, False],
    'copy_X': [True, False],
    'n_jobs': [None, -1]
}
```

```python
#tune random forest regressor

param_grid={
    'n_estimators': [50, 100, 200],
    'max_depth': [None,3,5,10, 15,20, 30],
    'min_samples_split': [1, 2, 5, 10],
    'min_samples_leaf': [1, 2, 3,5,10]
}

#tune
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(x_train, y_train)
best_params = grid_search.best_params_

print("Best parameters for Random Forest Regressor:", best_params)

#show metrics
best_rf_model = grid_search.best_estimator_
y_pred = best_rf_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
```
